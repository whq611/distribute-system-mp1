```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
lists = []
for i in range(3):
    with open("node" + str(i) + ".txt", "r") as fs:
        lists.append([])
        flag = False
        for line in fs:
            #print(line)
            if str(line) == "BANDWIDTH: \n":
                flag = True
                continue
            if line == "DELAY: \n":
                flag = False
            if flag:
                if line != "\n":
                    lists[i].append(float(line.split()[0]))
```

```python
a = np.array(lists)
plt.figure(figsize=(20,10))
for i in range(3):
    plt.plot(a[i]) 
plt.legend(['node0', 'node1', 'node2'])
plt.xlabel('time(s)')
plt.ylabel('Bandwidth(ms)')
plt.title('Evaluation bandwidth')
```
